```python
import tensorflow as tf
def one_hot_matrix(Y_train,C):
    C = tf.constant(C)
    one_hot_matrix = tf.one_hot(Y_train,C,axis=0)
    with tf.Session() as sess:
        init = tf.global_variables_initializer()
        sess.run(init)
        one_hot = sess.run(one_hot_matrix)
    return one_hot
```

```python
## Creating placeholder
def create_placeholder(n_x,n_y):
    X = tf.placeholder(dtype = tf.float32,shape=[n_x,None])
    Y = tf.placeholder(dtype = tf.float32,shape=[n_y,None])
    return X,Y
```

```python
## Initializing Parameters
def initialize_parameter():
    """
    Initializer parameter to build a neural network with tensorflow.
        Basically we are building a two hidden layer neural network
            W1 = [200,784]
            b1 = [200,1]
            W2 = [100,200]
            b2 = [100,1]
            W3 = [10,100]
            b3 = [10,1]
    """
    tf.set_random_seed(1)
    W1 = tf.get_variable("W1",[200,784],initializer=tf.contrib.layers.xavier_initializer(seed = 1))
    b1 = tf.get_variable("b1",[200,1],initializer =tf.zeros_initializer())
    W2 = tf.get_variable("W2",[100,200],initializer=tf.contrib.layers.xavier_initializer(seed = 1))
    b2 = tf.get_variable("b2",[100,1],initializer =tf.zeros_initializer())
    W3 = tf.get_variable("W3",[10,100],initializer=tf.contrib.layers.xavier_initializer(seed = 1))
    b3 = tf.get_variable("b3",[10,1],initializer =tf.zeros_initializer())
    
    parameters = {"W1":W1,
                  "b1":b1,
                  "W2":W2,
                  "b2":b2,
                  "W3":W3,
                  "b3":b3}
    return parameters
    
```

```python
tf.reset_default_graph()
with tf.Session() as sess:
    parameters = initialize_parameter()
    print("W1 = "+ str(parameters["W1"]))
    print("W2 = "+ str(parameters["W2"]))
    print("b1 = "+str(parameters["b1"]))
    print("b2 = "+str(parameters["b2"]))
```

```python
def forward_propagation(X,parameters):
    # Reteriving the parameters from dictionary "parameters"
    ## DropOut
    keep_prob = tf.constant(0.5)
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]
    b1 = parameters["b1"]
    b2 = parameters["b2"]
    b3 = parameters["b3"]
    # Forward propagation
    Z1 = tf.add(tf.matmul(W1,X),b1)
    A11 = tf.nn.relu(Z1)
    A1 = tf.nn.dropout(A11,keep_prob = keep_prob)
    Z2 = tf.add(tf.matmul(W2,A1),b2)
    A21 = tf.nn.relu(Z2)
    A2 = tf.nn.dropout(A21,keep_prob = keep_prob)
    Z31 = tf.add(tf.matmul(W3,A2),b3)
    Z3 = tf.nn.dropout(Z31,keep_prob = keep_prob)
    return Z3
```

```python
def compute_cost(Z3,Y):
    logits = tf.transpose(Z3)
    labels = tf.transpose(Y)
    cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits =logits,labels=labels))
    return cost
```

```python
tf.reset_default_graph()
with tf.Session() as sess:
    X,Y=create_placeholder(784,10)
    parameters = initialize_parameter()
    Z3 = forward_propagation(X,parameters)
    cost = compute_cost(Z3,Y)
    print("cost = "+str(cost))
```

```python
## MINIBATCH
def random_mini_batch(X,Y,mini_batch_size = 64,seed=0):
    np.random.seed(seed)
    m = X.shape[1]
    mini_batches = []
    # step 1 shuffle (X,Y)
    permutation = list(np.random.permutation(m))
    shuffled_X = X[:,permutation]
    shuffled_Y = Y[:,permutation]
    num_complete_minibatches = m//mini_batch_size
    for k in range(0,num_complete_minibatches):
        mini_batch_X = shuffled_X[:,k*mini_batch_size:(k+1)*mini_batch_size]
        mini_batch_Y = shuffled_Y[:,k*mini_batch_size:(k+1)*mini_batch_size]
        mini_batch= (mini_batch_X,mini_batch_Y)
        mini_batches.append(mini_batch)
    if not m%mini_batch_size==0:
        mini_batch_X = shuffled_X[:,(m - (num_complete_minibatches*mini_batch_size)):m]
        mini_batch_Y = shuffled_Y[:,(m - (num_complete_minibatches*mini_batch_size)):m]
        mini_batch = (mini_batch_X,mini_batch_Y)
        mini_batches.append(mini_batch)
        return mini_batches
```

```python
def model(X_train,Y_train,learning_rate = 0.001):
    tf.reset_default_graph()
    tf.set_random_seed(1)
    seed = 3
    (n_x,m) = train_X.shape
    n_y = train_Y.shape[0]
    costl = []
    ## Create placeholder
    X,Y= create_placeholder(n_x,n_y)
    ## INITIALIZE PARAMETERS
    parameters = initialize_parameter()
    ## forward propagation
    Z3 = forward_propagation(X,parameters)
    ## compute cost
    cost = compute_cost(Z3,Y)
    ## Backpropagation and optimizing weight in same step
    optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(cost)
    init = tf.global_variables_initializer()
    
    with tf.Session() as sess:
        sess.run(init)
        for i in range(1000):
            epoch_cost = 0
            seed = seed +1
            minibatches = random_mini_batch(train_X,train_Y,mini_batch_size=512)
            for minibatch in minibatches:
                minibatch_X,minibatch_Y = minibatch
                _ , costs = sess.run([optimizer,cost],feed_dict = {X:minibatch_X,Y:minibatch_Y})
                epoch_cost += costs/int(m/64)
            if i % 5 == 0:
                costl.append(epoch_cost)
            if i % 100 == 0:
                print("Cost after "+ str(i)+ " epochs = "+str(costs))
        # lets save the parmaeters
        parameters_final = sess.run(parameters)
        ## calculate prediction
        correct_prediction = tf.equal(tf.argmax(Z3),tf.argmax(Y))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
        
        print("Train Accuracy:",accuracy.eval({X:X_train,Y:Y_train}))
        return parameters_final,costl
```

```python
import time
tic = time.time()
parameters_final,costl = model(train_X,train_Y)
toc = time.time()
print("Time taken for with minibatch :"+str(1000*(toc-tic))+"ms")
```

```python
def forward_propagation_np(X,parameters):
    # Reteriving the parameters from dictionary "parameters"
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]
    b1 = parameters["b1"]
    b2 = parameters["b2"]
    b3 = parameters["b3"]
    # Forward propagation
    Z1 = np.add(tf.matmul(W1,X),b1)
    A1 = tf.nn.relu(Z1)
    Z2 = tf.add(tf.matmul(W2,A1),b2)
    A2 = tf.nn.relu(Z2)
    Z3 = tf.add(tf.matmul(W3,A2),b3)
    return Z3
```

```python
def test_result(X_test,parameters_final):
    n_x = test_X.shape[0]
    n_y = 24
    X,Y = create_placeholder(n_x,n_y)
    z3 = forward_propagation(X,parameters_final)
    with tf.Session() as sess:
        init = tf.global_variables_initializer()
        sess.run(init)
        z_3 = sess.run(z3,feed_dict = {X:X_test})
        return z_3
    
```

```python
z_3 = test_result(test_X,parameters_final)
```
